## Observations and Insights 

```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files

mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

study_data = pd.merge(mouse_metadata, study_results, on = 'Mouse ID')

# Display the data table for preview

study_data.head()
```

```python
# Checking the number of mice.

mouse_count = study_data['Mouse ID'].nunique()
mouse_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_mouse = study_data.loc[study_data.duplicated(subset = ["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
duplicate_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

duplicate_df = study_data.loc[study_data["Mouse ID"] == "g989"]
duplicate_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

clean_data=study_data[study_data["Mouse ID"].isin(duplicate_mouse) == False]
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.

mouse_count = clean_data['Mouse ID'].nunique()
mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.

mean = clean_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = clean_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = clean_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
standard_dev = clean_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = clean_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary dataframe.

summary_stats = pd.DataFrame({"Mean": mean, "Median": median, "Variance": variance, "Standard Deviation": standard_dev, "SEM": sem})
summary_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

stats = clean_data.groupby("Drug Regimen").agg({'Tumor Volume (mm3)' : ['mean', 'median', 'var', 'std', 'sem']})
stats


```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pandas.

drug_count = clean_data["Drug Regimen"].value_counts()
y_axis = drug_count.values
x_axis = drug_count.index

drug_count.plot(kind = "bar", facecolor = 'blue',)

plt.ylabel("Treatment Count")
plt.xlabel("Drug Regimen")
plt.title("Unique Mice Tested")
plt.xticks(rotation = 90)
plt.show()
```

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pyplot.

drug_count = clean_data["Drug Regimen"].value_counts()

y_axis = drug_count.values
x_axis = np.arange(0,len(y_axis),1)
x_drugs = drug_count.index

plt.bar(x_drugs, y_axis, color = 'b', alpha = 0.5, align = "center", width = .5)

plt.ylabel("Treatment Count")
plt.xlabel("Drug Regimen")
plt.title("Unique Mice Tested")
plt.xticks(rotation = 90)

plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas

mice_count = clean_data["Sex"].value_counts()

labels = mice_count.values

sizes = mice_count.values

colors = ["b", "r"]

explode = [0.1,0]

mice_count.plot(kind = "pie", explode = explode, colors = colors, autopct = "%1.1f%%", shadow = True, startangle = 90)

plt.title("Mouse Gender Comparison")
plt.ylabel(" ")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot

gender_count = (clean_data.groupby(["Sex"])["Age_months"].count()).tolist()

labels = ["Female", "Males"]
colors = ["blue", "red"]
explode = (0.1, 0)

plt.pie(gender_count, explode=explode, labels=labels, colors = colors, autopct = "%1.1f%%", shadow = True, startangle = 90)
plt.axis("equal")
plt.title("Mouse Gender Comparison")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

capomulin_data = clean_data.loc[clean_data["Drug Regimen"] == "Capomulin"]
capomulin_data = capomulin_data.reset_index()

capomulin_mice = capomulin_data.loc[capomulin_data["Mouse ID"] == "s185"]
capomulin_mice = capomulin_mice.loc[:, ["Timepoint", "Tumor Volume (mm3)"]]

capomulin_mice = capomulin_mice.reset_index(drop=True)
capomulin_mice.set_index("Timepoint").plot(figsize=(10,5), linewidth=3.0, color="red")
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

weight_df = capomulin_data.loc[:, ["Mouse ID", "Weight (g)", "Tumor Volume (mm3)"]]
average_cap_tumor_vol = pd.DataFrame(weight_df.groupby(["Mouse ID", "Weight (g)"])["Tumor Volume (mm3)"].mean()).reset_index()
average_cap_tumor_vol = average_cap_tumor_vol.rename(columns={"Tumor Volume (mm3)": "Average Tumor Volume"})
average_cap_tumor_vol.plot(kind="scatter", x="Weight (g)", y="Average Tumor Volume", grid=True, figsize=(5,5), title="Mouse Weight vs. Average Tumor Volume")
plt.show()

```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

from scipy.stats import linregress

mouse_weight = average_cap_tumor_vol.iloc[:,0]
average_tumor_volume = average_cap_tumor_vol.iloc[:,1]

x_values = average_cap_tumor_vol["Weight (g)"]
y_values = average_cap_tumor_vol["Average Tumor Volume"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values, y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,35),fontsize=15,color="red")
plt.xlabel("Mouse Weight")
plt.ylabel("Average Tumor Volume")
plt.show()
```
